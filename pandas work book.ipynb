```python
# multi index operation
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3] 
hire_index = list(zip(outside,inside))
hire_index = pd.MultiIndex.from_tuples(hire_index)
```

```python
hire_index
```

```python
df = pd.DataFrame(np.random.randn(6,3),index = hire_index,columns = ('A B C'.split()))
df
```
